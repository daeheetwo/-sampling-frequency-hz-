어제는 FFT 분석에 대해 알아봤는데 오늘은 주파수 성분만 보는 것이 아닌

시간성분과 함께 볼 수 있는 SPECTROGRAM 기법의 STFT 방법을 알아보고

이를 위한 WINDOW 를 확인할 수 있도록 하겠습니다.

먼저 간단한 실습을 위한 기초 데이터를 생성하도록 하겠습니다.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sin_wave(amp, freq, time): #Sin 함수 만들기
    return amp * np.sin(2*np.pi*freq*time)
```

```python
time = np.arange(0, 10, 0.001) #0~10초간 0.001 단위로 쪼개서 Time array 만들기
                               #np.arange(0, 10, 0.001) 의 경우 start=0, stop=10, step=0.001
sin1 = sin_wave(1, 10, time) #Sin wave 함수를 호출하여 1 amp / 10 freq / time 만큼 생성
sin2 = sin_wave(2, 5, time)  #위와 동일 Sin wave 함수를 호출
sin3 = sin_wave(4, 1, time)  #위와 동일 Sin wave 함수를 호출
```

```python
plt.figure(figsize=(12,5))
plt.plot(time, sin1, label=r"$\sin {20\pi} t$", color='red')
plt.plot(time, sin2, label=r"$2\sin {10\pi} t$", color='blue')
plt.plot(time, sin3, label=r"$4\sin {2\pi} t$", color='green')
plt.legend(); plt.grid(); plt.show()
```

```python
sin_sum = sin1 + sin2 + sin3
sin_concat = np.concatenate((sin1, sin2, sin3, sin_sum))
```

```python
time = np.arange(0, 40, 0.001)# 40000 개
plt.figure(figsize=(12,5))
plt.plot(time, sin_concat)
plt.grid()
plt.show()
```

```python
n = len(sin_concat)
k = np.arange(n)
Fs = 1/0.001; T = n/Fs
freq = k/T
```

```python
fft_values = np.fft.fft(sin_concat)  # FFT 수행
single_side_fft = np.abs(fft_values[:n // 2]) /n  # 양측 주파수 스펙트럼 계산

# 주파수 스펙트럼 그리기
plt.plot(freq[:n // 2], single_side_fft)
plt.title('Single Side FFT')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.show()
```

어제 여기까지 진행 했었습니다. 주파수 성분만으로는 시간 성분을 확인하기 어렵죠.

이를 해결하기 위한 방법론이 SPECTOGRAM 기법입니다.

x축은 시간, y축은 주파수를 보면 이미지 처럼 보이게 되는 것이지요.

image.png

가장 중요한 설명은 바로 SIGNAL 에 대한 WINDOW 와 OVERLAP 개념 입니다.

모든 시그널을 WINDOW 단위마다 짤라서 FFT 분석을 취합니다.

근데 이 시그널을 OVERLAP 만큼 겹쳐서 합니다.

그래서

Window Length / Overlap Length 가 STFT 분석에선 매우 중요합니다.

함께 확인해보겠습니다.

```python
def draw_stft(f, t, Zxx):
    plt.figure(figsize=(12,5))
    plt.pcolormesh(t, f, np.abs(Zxx), vmin=0, vmax=1, shading='gouraud')  #PCOLORMESH 라는 기능을 이용한 그림 그리기
    plt.title('STFT Magnitude'); plt.ylabel('Frequency [Hz]')
    plt.xlabel('Time [sec]'); plt.ylim([0, 20]); plt.show()
```

```python
from scipy import signal

def calc_stft(nperseg):
    f, t, Zxx = signal.stft(sin_concat, Fs, nperseg=nperseg)  # 매우 간단한 한 줄로 STFT 이미지를 제공할 수 있는 SCIPY SIGNAL 모듈
    draw_stft(f, t, Zxx)
```

```python
calc_stft(500)
```

https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.stft.html#scipy.signal.stft

공식 사이트를 한번 가보시면 다양한 기능들이 존재하고 있습니다.

image.png

우리가 알아야 하는 것은
1. Window opt = 'hann'   hann 이란 것은 무엇인지
2. nperseg opt = 256     256 의 의미는 무엇인지
3. noverlap = None       None 의 의미는 무엇인지 확인하는 것입니다.

옵션을 한번 줘봅시다 'hann'

```python
def calc_stft(nperseg):
    f, t, Zxx = signal.stft(sin_concat, Fs, window='hann', nperseg=nperseg)  # 매우 간단한 한 줄로 STFT 이미지를 제공할 수 있는 SCIPY SIGNAL 모듈
    draw_stft(f, t, Zxx)
```

```python
calc_stft(500)
```

```python
calc_stft(5000)
```

윈도우 크기를 조정함으로서 데이터의 구분이 명확하게 보이게 됩니다.


사실 이게 중요한 사유는 우리가 나중에 이 이미지로 모델링을 할 것이기 때문입니다.


다음은 scipy 에서 제공하는 윈도우 종류입니다. 아무거나 위 코드에 대입해서 적용해보십시요 그리고 다름을 확인하십시요

boxcar, triang, blackman, hamming, hann, bartlett, flattop, parzen, bohman,blackmanharris, nuttall, barthann, cosine, exponential, tukey, taylor, lanczos, gaussian, general_cosine, general_gaussian, general_hamming, dpss

```python
def calc_normal_stft(nperseg):
    f, t, Zxx = signal.stft(sin_concat, Fs, nperseg=nperseg)  # 매우 간단한 한 줄로 STFT 이미지를 제공할 수 있는 SCIPY SIGNAL 모듈
    draw_stft(f, t, Zxx)
def calc_window_stft(nperseg):
    f, t, Zxx = signal.stft(sin_concat, Fs, window='hamming' , nperseg=nperseg)  # hamming 부분을 다른 윈도루로 변경할 것
    draw_stft(f, t, Zxx)
```

```python
calc_normal_stft(500)
calc_window_stft(500)
```

이번에는 overlap 입니다. 얼마나 신호를 겹치게 만들게 되면 될까요?

일반적으로 Overlap 퍼센트는 50~75%사이로 사용합니다. 이는 분석하여 잘 보이는 이미지를 찾는 과정입니다

```python
def calc_overlap_stft(nperseg):
    f, t, Zxx = signal.stft(sin_concat, Fs, nperseg=nperseg, noverlap=nperseg//3)  # 숫자를 바꾸면 overlap 이 바뀝니다. 3이면 1/3 겹친다는 의미죠
    draw_stft(f, t, Zxx)
```

```python
calc_overlap_stft(500)
```
