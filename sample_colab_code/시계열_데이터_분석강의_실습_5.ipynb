## Breif Description of Test Rig for Data Collection

This is a <em><b>"Run to Failure "</em></b> dataset.

Four bearings were installed on a shaft. The rotation speed was kept constant at 2000 RPM by an AC motor coupled to the shaft via rub belts.  A radial load of 6000 lbs is applied onto the shaft and bearing by a spring mechanism. All bearings are force lubricated.

Rexnord ZA-2115 double row bearings were installed on the shaft as shown in Figure 1. PCB 353B33 High Sensitivity Quartz ICP accelerometers were installed on the bearing housing (two accelerometers for each bearing [x- and y-axes] for data set 1, one accelerometer for each bearing for data sets 2 and 3). Sensor placement is also shown in figure. All failures occurred after exceeding designed life time of the bearing which is more than 100 million revolutions.

```python
healthy_bearing1 = merged_data['2004-02-12 10:32:39':'2004-02-15 23:42:39']['Bearing 1']
faulty_bearing1 = merged_data['2004-02-18 11:02:39':'2004-02-18 23:52:39']['Bearing 1']
```

```python
sample_rate = 34.1
```

```python
# 푸리에 변환을 위한 함수
def perform_fft(signal):
    fft_result = np.fft.fft(signal)
    magnitude = np.abs(fft_result)
    fs = sample_rate
    nfft = len(signal)
    sr = fs / nfft
    f = np.linspace(0, sr, len(magnitude))
    left_spectrum = magnitude[:int(len(magnitude) / 2)]
    left_f = f[:int(len(magnitude) / 2)]
    return left_f, left_spectrum

# 정상 상태와 고장 상태의 FFT 분석
f_healthy, spectrum_healthy = perform_fft(healthy_bearing1)
f_faulty, spectrum_faulty = perform_fft(faulty_bearing1)

# 시각화
plt.figure(figsize=(20, 5))
plt.title("Healthy vs Faulty Bearing 1")
plt.plot(f_healthy, spectrum_healthy, label='Healthy')
plt.plot(f_faulty, spectrum_faulty, label='Faulty')
plt.xlabel("Frequency")
plt.ylabel("Magnitude")
plt.legend()
plt.show()
```

```python
# STFT 변환을 위한 함수
from scipy.signal import stft

f_healthy, spectrum_healthy = perform_fft(fft_dataset_normal.iloc[:, 0])
f_faulty, spectrum_faulty = perform_fft(fft_dataset_abnormal.iloc[:, 0])

#fft_dataset_normal=pd.read_csv('/content/2nd_test/2004.02.12.10.32.39', sep='\t', header=None)
#fft_dataset_abnormal=pd.read_csv('/content/2nd_test/2004.02.18.11.02.39', sep='\t', header=None)

fs = 34.1
nperseg = 256
noverlap = nperseg // 2
nfft = nperseg

# STFT 변환을 위한 함수
def perform_stft(signal):
    f, t, Zxx = stft(signal, fs=fs, nperseg=nperseg, noverlap=noverlap, nfft=nfft)
    return f, t, Zxx

# STFT 변환 수행
f, t, Zxx = perform_stft(fft_dataset_normal.iloc[:, 0])

# 그래프 생성 및 저장
plt.figure()
plt.pcolormesh(t, f, np.abs(Zxx), vmin=0, vmax=np.abs(Zxx).max(), shading='auto', cmap='jet')
output_normal_file_name = '2004.02.12.10.32.39' + '_stft.png'
plt.savefig(output_normal_file_name)

# STFT 변환 수행
f, t, Zxx = perform_stft(fft_dataset_abnormal.iloc[:, 0])

# 그래프 생성 및 저장
plt.figure()
plt.pcolormesh(t, f, np.abs(Zxx), vmin=0, vmax=np.abs(Zxx).max(), shading='auto', cmap='jet')
output_abnormal_file_name = '2004.02.18.11.02.39' + '_stft.png'
plt.savefig(output_abnormal_file_name)

from IPython.display import Image
Image(output_normal_file_name), Image(output_abnormal_file_name)
```

```python


# Wavelet 변환을 위한 함수
def perform_wavelet(signal, wavelet='shan'):#특이파형 적용
    scales = np.arange(1, 128)
    coefficients, frequencies = pywt.cwt(signal, scales, wavelet)
    return np.abs(coefficients), scales

# Wavelet 변환 수행
coefficients_healthy, scales_healthy = perform_wavelet(fft_dataset_normal.iloc[:, 0])
coefficients_faulty, scales_faulty = perform_wavelet(fft_dataset_abnormal.iloc[:, 0])

# 그래프 생성 및 저장
plt.figure(figsize=(15, 5))
plt.imshow(coefficients_healthy, aspect='auto', extent=[0, len(coefficients_healthy[0]), scales_healthy[-1], scales_healthy[0]], cmap='jet')
plt.title("Wavelet Transform - Healthy")
plt.xlabel("Time")
plt.ylabel("Scale")
plt.colorbar()
plt.savefig('wavelet_healthy.png')
plt.show()

# 그래프 생성 및 저장
plt.figure(figsize=(15, 5))
plt.imshow(coefficients_faulty, aspect='auto', extent=[0, len(coefficients_faulty[0]), scales_faulty[-1], scales_faulty[0]], cmap='jet')
plt.title("Wavelet Transform - Faulty")
plt.xlabel("Time")
plt.ylabel("Scale")
plt.colorbar()
plt.savefig('wavelet_faulty.png')
plt.show()

# 이미지 출력
from IPython.display import Image
Image('wavelet_healthy.png'), Image('wavelet_faulty.png')
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Flatten, Reshape
from sklearn.model_selection import train_test_split

# Function to load and preprocess the image data
def load_images(folder_path):
    file_list = os.listdir(folder_path)
    images = []
    labels = []  # Added labels for normal (0) and abnormal (1)

    for file_name in file_list:
        if file_name.endswith('.png'):
            file_path = os.path.join(folder_path, file_name)
            image = plt.imread(file_path)
            images.append(image)

            # Use the file name to determine if it's normal or abnormal
            label = 0 if '2004.02.18' or '2004.02.19' not in file_name else 1
            labels.append(label)

    return np.array(images), np.array(labels)

# Load image dataset with labels
img_folder = '/content/imgdataset'
images, labels = load_images(img_folder)

# Normalize images
images = images / 255.0

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(images, labels, test_size=0.2, random_state=42)

# Define the autoencoder model
input_shape = (X_train.shape[1], X_train.shape[2], X_train.shape[3])
latent_dim = 32  # Adjust as needed

input_layer = Input(shape=input_shape)
flatten_layer = Flatten()(input_layer)
encoded_layer = Dense(latent_dim, activation='relu')(flatten_layer)
decoded_layer = Dense(np.prod(input_shape), activation='sigmoid')(encoded_layer)
output_layer = Reshape(input_shape)(decoded_layer)

autoencoder = Model(input_layer, output_layer)
autoencoder.compile(optimizer='adam', loss='mse')

# Train the autoencoder on normal data
autoencoder.fit(X_train[y_train == 0], X_train[y_train == 0], epochs=50, batch_size=4, validation_data=(X_test[y_test == 0], X_test[y_test == 0]))

# Evaluate the model on the test set
losses = autoencoder.evaluate(X_test, X_test, verbose=0)
print(f"Mean Squared Error on Test Set: {losses}")

# Generate predictions on the entire dataset
predictions = autoencoder.predict(images)

# Calculate the reconstruction errors
reconstruction_errors = np.mean(np.square(images - predictions), axis=(1, 2, 3))

# Set a threshold for anomaly detection (you may need to adjust this)
threshold = 0.02

# Identify anomalies based on the threshold
anomalies = reconstruction_errors > threshold

# Display anomalies
anomaly_images = images[anomalies]
for i, anomaly in enumerate(anomaly_images):
    plt.figure(figsize=(4, 4))
    plt.imshow(anomaly)
    plt.title(f"Anomaly {i + 1}")
    plt.show()
```
